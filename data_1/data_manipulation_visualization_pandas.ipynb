```python
# Example, DataFrame with 3 columns

import numpy as np
import pandas as pd
t = np.arange(0,10,0.1)
x = np.sin(t)
y = np.cos(t)
df = pd.DataFrame({'Time':t, 'x':x, 'y':y})
```

```python
#More complex manipulations are possible
# E.g. grouping objects and performing statistical operations on groups: 

import matplotlib.pyplot as plt
data = pd.DataFrame({
'Gender': ['f', 'f', 'm', 'f', 'm',
'm', 'f', 'm', 'f', 'm', 'm'],
'TV': [3.4, 3.5, 2.6, 4.7, 4.1, 4.1,
5.1, 3.9, 3.7, 2.1, 4.3]
})

# Grouping:
grouped = data.groupby('Gender')
#NaN group also exists

# Some basic stats:
print(grouped.describe())
# ploting:
grouped.boxplot()
plt.show()
```

```python
#  DataFrame object based on grouping:
df_female = grouped.get_group('f')
# numpy array creating:
values_female = grouped.get_group('f').values
print(df_female)
print(values_female)
```

```python
# Line fitting

x = np.random.randn(200)
y = 10+0.5*x+np.random.randn(len(x))

# Scatter plot
plt.scatter(x,y)
plt.title('Scatter plot of data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
M = np.vstack((np.ones(len(x)), x)).T
pars = np.linalg.lstsq(M,y)[0]
a = pars[0]
b = pars[1]
plt.scatter(x,y)
plt.plot(x, a + b*x, 'r')
plt.show()
```

```python
# Exercise:
#1. Read files heroes_information.csv , super_hero_powers.csv
#2. Peak into DataFrames - head(), tail()
#3. Show column with names of superheroes
#4. List all powers (all column names, except first in super_hero_powers.csv), use method columns()
#5. Find number of good heroes, use value_counts()
#6. Find 5 most frequent powers 
#7. Find 'unique' powers (powers that only one hero has)
#8. Using bar plot show the number of heros of each Publisher
#9. Show histogram of the heights of the heroes
```

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 2: Read the CSV files
heroes_info = pd.read_csv('heroes_information.csv')
hero_powers = pd.read_csv('super_hero_powers.csv')
```

```python
# Step 3: Peek into the DataFrames
print("Heroes Info - Head:")
display(heroes_info.head())

print("Heroes Info - Tail:")
display(heroes_info.tail())

print("Hero Powers - Head:")
display(hero_powers.head())

print("Hero Powers - Head:")
display(hero_powers.tail())
```

```python
# Step 4: Show column with superhero names
print("Column with superhero names:")
print(heroes_info['name'].head())   # shows first few names
```

```python
# Step 5: List all powers (all column names except the first one)
powers = hero_powers.columns[1:]
print("List of powers:")
print(powers)
```

```python
# Step 6: Find number of good heroes (use value_counts on Alignment column)
print("Hero alignment counts:")
print(heroes_info['Alignment'].value_counts())

num_good = heroes_info['Alignment'].value_counts().get('good', 0)
print(f"\nNumber of good heroes: {num_good}")
```

```python

# Step 7: Find 5 most frequent powers
# Convert True/False to count per column
power_counts = hero_powers.iloc[:, 1:].sum().sort_values(ascending=False)
print("\nTop 5 most frequent powers:")
print(power_counts.head(5))
```

```python
# Step 8: Find 'unique' powers (only one hero has True)
unique_powers = power_counts[power_counts == 1]
print("\nPowers that only one hero has:")
print(unique_powers)
```

```python
# Step 9: Bar plot – number of heroes per Publisher
publisher_counts = heroes_info['Publisher'].value_counts().dropna()
publisher_counts.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('Number of Heroes by Publisher')
plt.xlabel('Publisher')
plt.ylabel('Number of Heroes')
plt.show()
```

```python
# Step 10: Histogram – distribution of hero heights
# Clean the 'Height' column (remove NaN and zeros)
heights = heroes_info['Height'].dropna()
heights = heights[heights > 0]

plt.figure(figsize=(10,6))
plt.hist(heights, bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Hero Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Number of Heroes')
plt.show()
```
