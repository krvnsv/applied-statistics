```python
red = ['r'] * 8
blue = ['b'] * 6
```

```python
box = red + blue
box
```

```python
import numpy as np
```

```python
np.random.shuffle(box)
```

```python
box
```

```python
a = np.random.choice(len(box), size=2, replace=False)
a
```

```python
box[a[0]] == 'r'
```

```python
N = 100000
P = []
favorable_outcome = 0

for i in range(1, N+1):
  # draw from the box
  indices = np.random.choice(len(box), size=2, replace=False)
  if ((box[indices[0]] == 'r') and (box[indices[1]] == 'b')) or ((box[indices[0]] == 'b') and (box[indices[1]] == 'r')):
    favorable_outcome += 1
  P.append(favorable_outcome/i)
print(favorable_outcome/i)
```

```python
t_P = 48/91
t_P
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(np.arange(1, N+1), P)
plt.plot(np.arange(N), [t_P] * N)
```
